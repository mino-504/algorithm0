# 문제정의
정렬되지 않은 리스트의 탐색 문제


# 알고리즘 설명
리스트에 n개의 항목이 들어있을때 이 리스트에서 탐색키를 가진 항목을 찾기

만약 찾는 항목이 있다면 그 항목의 인덱스를 반환, 없으면 -1을 반환

리스트의 항목은 정렬되있지 않음

# 손으로 푼 예제


# 코드 개요
입력 변수 : A, key

출력 : arr,key

코드 개요 : 리스트의 모든 값을 순회하며 탐색한 뒤, key의 변수에 해당하는 값이 있다면 그 값은 반환하고 key의 변수에 해당하는 값이 없을경우 -1을 반환한다.

```python
# 알고리즘 코드
def sequential_search(A,key):
    for i in range(len(A)):
        if A[i]==key:
            return i
    return -1
```

```python
# 테스트 코드
arr=[7,4,2,1,6,9,8]
key=9
print("Original: ",arr)
sequential_search(arr,key)
print("Selection: ",key)
```

# 수행결과



# 복잡도 분석
최선의 경우 : 리스트의 첫 번째 항목에 key 값이 존재하는 경우 : O(1)

평균의 경우 : 리스트의 모든 값이 key 값인 경우 : (n+1)/2 = O(n)

최악의 경우 : 리스트의 마지막 항목에 key가 존재하는 경우, 혹은 key 값이 존재하지 않는 경우 : n = O(n)
