# 문제 정의
문자열 매칭 문제

# 알고리즘 설명
길이가 n인 입력 문자열 T와 길이가 m인 패턴 문자열 P가 있을 떄, T에서 가장 먼저 나타나는 P의 위치 찾기.

(없으면 -1을 반환)

# 손으로 푼 예제


# 코드 개요
입력변수 : 입력 문자열(T), 패턴 문자열(P)

출력 : 입력 문자열(T), 패턴 문자열(P), 함숫값(string_matching)

코드 개요 : 패턴의 문자들을 입력 문자열의 모든 값을 순회하며 탐색한 뒤, 패턴의 문자들이 일치하면 그 위치를 반환하고, 문자들이 일치하지 않을 경우, -1을 반환한다.

```python
#알고리즘 코드
def string_matching(T,P):
    n=len(T)
    m=len(P)
    for i in range(n-m+1):
        j=0
        while j<m and P[j]==T[i+j]:
            j=j+1
        if j==m:
            return i
    return -1
```

```python
#테스트 코드
text="ALGORITHM IS FUN"
pattern ="S F"
print(pattern,'in',text,'-->',string_matching(text,pattern))
pattern="GR"
print(pattern,'in',text,'-->',string_matching(text,pattern))
```

# 수행 결과


# 복잡도 분석
최선의 경우 : 패턴 문자열(P)이 입력문자열(T) 바로 앞에 존재하는 경우, 패턴 문자열의 길이(m)만큼 비교하면 되므로 O(m)이다.

최악의 경우 : 패턴 문자열(P)이 입력문자열(T)의 마지막까지 검사하는 경우, 입력문자열의 길이(n) - 패턴문자열의 길이(m)+1번  X 패턴 문자열의 길이(m) 만큼 비교해야 하므로 O(mn)이다.
