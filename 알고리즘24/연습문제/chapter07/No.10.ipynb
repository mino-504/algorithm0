### No.10

>>DP 테이블의 크기: 무게가 실수 단위라면 DP 테이블의 크기가 무한해질 수 있다. 예를 들어, 무게의 범위가 [0, 1]이고 소수점 이하 n 자리까지 고려해야 한다면, DP 테이블의 크기는 10^n 만큼 커진다.   
>
>>인덱싱의 어려움: 실수를 인덱스로 사용하는 것은 불가능하므로, 무게를 정수 인덱스로 맵핑하는 과정이 필요하다. 이는 실수 단위의 무게를 적절한 정수로 변환해야 하며, 변환 과정에서의 정밀도 문제와 오차가 발생할 수 있다.   
>
>>시간 및 공간 복잡도 증가: 소수점 이하의 자리수를 늘릴수록 DP 테이블의 크기와 연산 복잡도가 기하급수적으로 증가한다. 예를 들어, 소수점 이하 3자리까지 고려하면, DP 테이블의 크기는 1000배 증가합니다. 이는 시간 복잡도와 공간 복잡도를 크게 증가시키며 실질적으로 계산이 불가능해질 수 있다.
