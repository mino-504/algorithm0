>이진트리의 높이

>#### **2. 알고리즘 설명(MD)  (알고리즘 설명)**
>이진트리가 공백이면 0을 반환하며 최고 수를 구한다

#### **3. 손으로 푼 예제 (![이름][img]/이미지.jpg])  (책 예제 코드 이해 후 사진)**

>#### **4. 코드 개요(입력 변수, 출력, 알고리즘 개요) (MD)**
>##### (코드 개요 설명 1. 입력 변수 2. 출력 3. 함수설명)
>1. 입력 변수 : calc_height(root) : 
>2. 출력      &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;: 호출문을 호출하게 되면 트리의 가장 높은 곳을 찾아서 높이가 몇인지 알려준다.
>3. 함수 설명 : 왼쪽 트리의 가장 높은 곳까지 쭉 가서 더이상 자식 노드가 없다면 0을 반환해 모든곳을 둘러보고 첫 왼쪽과 오른쪽의 높이를 비교해서 제일 높은 곳의 수를 반환해준다

---------

#### **5. 코드(PY, C)**

```python
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft,hRight) + 1

d = TNode('D', None, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', None, None)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(" 트리의 높이 = ", calc_height(root))
```

------

#### **6. 테스트 코드(PY, C), 입력을 바꾸어서**

```python
class TNode:
    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right

def calc_height(root) :
    if root is None :
        return 0
    hLeft = calc_height(root.left)
    hRight = calc_height(root.right)
    return max(hLeft,hRight) + 1

z = TNode('Z', None, None)
i = TNode('I', None, z)
h = TNode('H', None, None)
p = TNode('P', None, None)
g = TNode('G', None, p)
d = TNode('D', g, None)
e = TNode('E', None, None)
b = TNode('B', d, e)
f = TNode('F', h, i)
c = TNode('C', f, None)
root = TNode('A', b, c)

print(" 트리의 높이 = ", calc_height(root))
```
