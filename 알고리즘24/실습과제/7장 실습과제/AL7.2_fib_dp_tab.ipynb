#### **5. 코드(PY, C)**

```python
def fib_dp_tab(n):
    f = [None] * (n + 1)
    f[0] = 0
    f[1] = 1
    for i in range(2, n + 1):
        f[i]=f[i-1] + f[i-2]
    return f[n]
    
n = 8
print('동적계획(테이블화): Fibonacci(%d) = '% n, fib_dp_tab(n))
```

#### **6. 테스트 코드(PY, C), 입력을 바꾸어서**

```python
def fib_dp_tab(n):
    f = [None] * (n + 1)
    f[0] = 0
    f[1] = 1
    for i in range(2, n + 1):
        f[i]=f[i-1] + f[i-2]
    return f[n]
    
n = 13
print('동적계획(테이블화): Fibonacci(%d) = '% n, fib_dp_tab(n))
```
